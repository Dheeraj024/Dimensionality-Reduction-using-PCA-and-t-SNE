```python
#import data
from keras.datasets import mnist
```

```python
(X,y),(_,_) = mnist.load_data()
print(X.shape, y.shape)
```

```python
X = X.reshape(len(X), -1)
X.shape
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
import matplotlib.pyplot as plt
```

```python
##pca
x_pca = PCA(n_components=2).fit_transform(X)
# sc = plt.figure(figsize = (10,10))
sc = plt.scatter(x_pca[:, 0],x_pca[:,1], c= y)
plt.legend(handles = sc.legend_elements()[0], labels = list(range(10)))
plt.show()
```

most of the digits here are overlapping or are group together.

```python
# t-SNE
x_tsne = TSNE(n_jobs = -1).fit_transform(X[:10000])
```

```python
plt.figure(figsize=(10,10))
sc = plt.scatter(x_tsne[:, 0],x_tsne[:,1], c= y[:10000])
plt.legend(handles = sc.legend_elements()[0], labels = list(range(10)))
plt.show()
```

Here using t-SNE approach we can see better visualization as compared to PCA.
